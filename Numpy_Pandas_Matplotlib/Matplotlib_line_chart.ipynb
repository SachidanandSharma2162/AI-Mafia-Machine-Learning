```python
# act as a graph paper
import matplotlib.pyplot as plt
```

```python
 # specify data points
x=[1,2,3]
y=[2,4,1]
```

```python
# draw thw graph
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('hours')
plt.ylabel('work')
plt.title('work vs hours graph')
plt.show()
```

```python
x1=[1,2,3]
y1=[2,4,6]

x2=[1,4,6,8]
y2=[3,1,7,3]
```

```python
plt.xlim(1,10)
plt.ylim(1,10)
plt.style.use('fivethirtyeight')
plt.plot(x1,y1,linestyle='dashed', marker='o', markerfacecolor='green', markersize=8 ,label='line-1')
plt.plot(x2,y2,linestyle='dashed', marker='o', markerfacecolor='black', markersize=8 ,label='line-2')
plt.legend()
plt.show()
```
